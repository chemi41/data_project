```python
# Когортный Анализ
```

```python
# Оставляем только те значения, которые нам нужны 
df = df[['client_id', 'id', 'created']].drop_duplicates()
```

```python
# Создаем 2 переменных: дата заказа и когорт
df['order_month'] = df['created'].dt.to_period('M')
df['cohort'] = df.groupby('client_id')['created'] \
                 .transform('min') \
                 .dt.to_period('M') 
```

```python
# Добавляем индиакторы для периодов(после первого месяца заказа)
df_cohort = df.groupby(['cohort', 'order_month']) \
              .agg(n_customers=('client_id', 'nunique')) \
              .reset_index(drop=False)
df_cohort['period_number'] = (df_cohort.order_month - df_cohort.cohort).apply(attrgetter('n'))
```

```python
# Превращаем данные в матрицу
cohort_pivot = df_cohort.pivot_table(index = 'cohort',
                                     columns = 'period_number',
                                     values = 'n_customers')
```

```python
# матрица удержания клиентов (размер когорта)
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis = 0)
```

```python
# график матрицы удерждания
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # матрица удержания
    sns.heatmap(retention_matrix, 
                mask=retention_matrix.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # размер когорта
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```
